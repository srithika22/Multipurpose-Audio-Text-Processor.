```python
from transformers import pipeline
```

```python
analyzer = pipeline('sentiment-analysis')
```

```python
generate = pipeline('text-generation')
```

```python
statement = input('Enter any short statement: ')
```

```python
result = generate(statement, max_length=150)
```
